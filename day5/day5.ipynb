If last number in rule is in the 1st list

Take two numbers at a time, from index 1 to len(rule) - 2

```python
import numpy as np

rule_list1 = []
rule_list2 = []
with open("rules.txt", "r") as file:
    for line in file:
        rule_list1.append(int(line.split("|")[0]))
        rule_list2.append(int(line.strip().split("|")[1]))
print(rule_list1)
print(rule_list2)
len_rules = len(rule_list1)
```

```python
updates = []
with open("updates.txt", "r") as file:
    for line in file:
        nums = line.strip().split(",")
        num_list = []
        for num in nums:
            num_list.append(int(num))
        updates.append(num_list)
print(updates)
```

```python
valid_updates=[]

def compare_nums(num1, num2):
    for j in range(0, len_rules):
        if rule_list1[j] == num2 and rule_list2[j] == num1:
            return False
    return True

for update in updates:
    len_update = len(update)

    valid_num_pair = []
    for i in range(0, len_update - 1):
        num1 = update[i]
        num2 = update[i + 1]
        valid_num_pair.append(compare_nums(num1, num2))

    if False in valid_num_pair:
        valid_updates.append(0)
    else:
        valid_updates.append(update[len_update//2])
        
print(sum(valid_updates))
```
